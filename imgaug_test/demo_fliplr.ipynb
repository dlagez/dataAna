```python
import imgaug.augmenters as iaa
import imageio
import imgaug as ia
```

```python
img = imageio.imread("/data/file/img/dog.jpg")
```

```python
print("original")
ia.imshow(img)
```

```python
aug = iaa.Flipud(0.5)
new_img = aug.augment_image(img)
ia.imshow(new_img)
```

```python
aug2 = iaa.Fliplr(0.6)
new_img2 = aug2.augment_image(img)
ia.imshow(new_img2)
```
